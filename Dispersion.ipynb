```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams['text.usetex'] = True
plt.rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
plt.rcParams["font.size"] = "12"
matplotlib.rcParams['axes.formatter.useoffset'] = False
```

```python
res = 600
x =   np.linspace(-np.pi,np.pi, res+1)
y =  2*(1-np.cos(x))+0.3

fig = plt.figure()
cm = 1 / 2.54
fig.set_size_inches(14.6 * cm, 8.6 * cm)


plt.xlabel("k*a")
plt.ylabel("E/(JS)")
plt.xlim(-np.pi,np.pi)
plt.ylim(0,4.4)

plt.grid(color='grey', linestyle='--', linewidth=0.5,zorder = -900)
plt.plot(x, y, label="Dispersion", color="red",linewidth=5.0)
ax = plt.gca()
ax.set_axisbelow(True)
plt.legend(labelspacing=0.19, handlelength=0.7, handletextpad=0.3, borderpad=0.2, loc="upper center", prop={'size': 12}, facecolor="white", framealpha=1)

plt.show()
```
